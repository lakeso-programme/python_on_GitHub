## Task 3
* Rank the Cost Effectiveness of the positions in England, and find the most "cost effective" position. The Cost Effectiveness is defined as wage/overall score. It reflects that the positions with less competitiveness but higher salary.



```python
# Task 3a
# Sort the English Players with overall and wage only

import pandas as pd
import numpy as np

filename = r'C:\Users\lakes\OneDrive\Documents\Lakeso\fifa_data.csv'
data = pd.read_csv(filename)

englandPlayers = data[(data['Nationality']=='England')][['Name','Overall','Wage','Position']]
print(englandPlayers)
```

```python
#Task 3b
# Add a column, CosT_Effectiveness, which is Wage / Overall

print(englandPlayers)
#But wage is string. Have to change it to float?

"""In pandas, we can use a function called split, very much like Excel LEFT RIGHT MID, \n
source: https://datatofish.com/left-right-mid-pandas/ \n
"""
df_no_euro = data['Wage'].str.split('€').str[1]
df_no_K = df_no_euro.str.replace('K','000',regex=True)



data['Wage1'] = df_no_K
data['Wage1'] = data['Wage1'].astype(int)
eng_df = data[(data['Nationality']=='England')][['ID','Name','Overall','Position','Wage','Wage1']]
print(eng_df)



#eng_str = d_no_euro.str.replace('K',' ',regex=True)
#eng_str
#eng_int = eng_str

```

```python
wageMedian = eng_df.groupby('Position').median()['Wage1']
scoreMedian = eng_df.groupby('Position').median()['Overall']

print(wageMedian)
print(scoreMedian)

result3 = pd.concat([wageMedian.rename('Wage'), scoreMedian.rename('Score')], axis=1)

# calculate the cost effective score
result3['CE'] = result3['Wage']/result3['Score']
result3
# normalize it for better look
result3['CE'] = (result3['CE']-result3['CE'].min())/(result3['CE'].max()-result3['CE'].min())

result3.sort_values(by=['CE'], ascending=False)
```
